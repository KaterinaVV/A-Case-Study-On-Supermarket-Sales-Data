```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('Article_top_Data.csv')
data.head()
```

```python
print(data.StoreNumber.unique())
encoder = LabelEncoder()
data.StoreNumber = encoder.fit_transform(data.StoreNumber)
print(data.StoreNumber.unique())
# data.StoreNumber = data.StoreNumber.apply(lambda x: f'store_{x}')
# print(data.StoreNumber.unique())
```

```python
sales = data.groupby(['Date','StoreNumber', 'CashRegisterNumber', 'ReceiptNumber','Group'], as_index=False).aggregate({'Quantity': 'sum'})
sales.head()
```

```python
# Feature matrix with shape (n_series x time_observations)
ts_series = sales.pivot(index=['Date','StoreNumber','CashRegisterNumber', 'ReceiptNumber'], columns=['Group'])
ts_series = ts_series.fillna(0)
ts_series.head(10)
```

```python
ts_series.reset_index(1, inplace=True)
ts_series.head()
```

```python
ts_series.columns = ts_series.columns.droplevel(0)
display(ts_series.head(2))
ts_series.rename(columns={'':'StoreNumber'}, inplace=True)
display(ts_series.head(2))
```

```python
ts_series.values
```

```python
ts_series.columns = ['StoreNumber','Augustiner','Avocado','BO-Laugenbreze','Bananen','Broetchen','GL_Sahne',
                 'Gurken','Kiwi_Gruen','Mozzarella','Milch','Pfandartikel','SammelNr','Schlaufentragetasche','VL_Eier','Zeitschriften']
ts_series.columns
```

```python
ts_series.columns
```

```python
ts_series.shape
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense

# 
sales_data = ts_series.to_numpy()
labels = np.array(ts_series.StoreNumber.values)
```

```python
# 
normalized_data = sales_data / np.max(sales_data)  # 
num_classes = len(np.unique(labels))  # 
print(normalized_data[:3])
print(labels[:3]) 
```

```python
# 
X = normalized_data[:, 1:]  # 
y = labels
```

```python
# LSTM[]
X = np.reshape(X, (X.shape[0], 1, X.shape[1]))
```

```python
# 
y = np.eye(num_classes)[y]
```

```python
# LSTM
model = Sequential()
model.add(LSTM(50, input_shape=(1, 15)))  # 50LSTM
model.add(Dense(num_classes, activation='softmax'))  # softmax
```

```python
# 
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 
model.fit(X, y, epochs=100, batch_size=1, verbose=2)
```

```python
# 
new_input = np.array([[5, 3]]) / np.max(sales_data)  # 
new_input = np.reshape(new_input, (new_input.shape[0], 1, new_input.shape[1]))  # 
predicted_output = model.predict(new_input)

predicted_label = np.argmax(predicted_output)

print(":", predicted_label)
```
